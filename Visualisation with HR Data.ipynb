## Veri Görselleştirme
Veri görselleştirme, verilerin anlaşılmasını kolaylaştıran ve bilgilerin etkili bir şekilde iletilmesini sağlayan önemli bir araçtır. Veri görselleştirmeleri, veri setindeki desenleri, ilişkileri, eğilimleri ve anormallikleri görsel olarak ortaya koyar. Grafikler, veri analistlerinin veri setindeki önemli noktaları ve potansiyel sorunları daha hızlı bir şekilde tespit etmelerine yardımcı olur.Görselleştirmeler, iş kararlarına ve stratejik planlamaya yardımcı olur. Verilerin görsel olarak sunulması, karar vericilerin trendleri, ilişkileri ve fırsatları daha iyi anlamalarını sağlar. Buna dayanarak, daha bilinçli ve bilgiye dayalı kararlar alınabilir.<br> <br>
Grafikler, veri setinin genel yapısını ve dağılımını daha kolay anlama ve anlama imkanı sağlar. Büyük veri setleriyle çalışırken, grafikler yardımıyla önemli noktaları vurgulayabilir, verileri daha kolay anlaşılır bir şekilde özetleyebilir ve genel bir bakış sunabilirsiniz. <br> <br>

Görselleştirmeler, performans göstergelerini izlemek ve takip etmek için de kullanılabilir. Örneğin, zamanla değişen verileri görsel olarak temsil ederek, bir işletmenin performansını, hedeflere ulaşma ilerlemesini veya trendleri izleyebilirsiniz.

Boş verimiz çıkmadığı için analiz ve görselleştirme aşamamıza geçebiliriz.Eğer boş verilerimiz çıksaydı bunlarla baş etme stratejimizi belirleyecektik. <br> <br>
Elimizdeki veri setinde her satır bir çalışanın verilerini içermektedir. Çalışanın bölümü, aldığı maaşın kategorisi (yüksek, düşük vb...), çalışma yılı, tatmin puanı gibi verilerimizi görselleştirerek elimizdeki İnsan Kaynakları verisine tek bakışta anlaşılabilecek analizlerle değer katacağız. <br>
* İlk olarak departman bazlı bazı grafik analizleri yapalım. Bunun için ana verimiz içinden yeni bir veri seti düzenleyelim:

```python
df_departments = df.groupby(by = 'Department', axis = 0).mean()
df_departments['number_of_emp'] = df.groupby(by = 'Department', axis = 0)['salary'].count()
df_departments.drop(['last_evaluation', 'Work_accident'], axis = 1, inplace = True)
df_departments.sort_values(by='number_of_emp', ascending = False)
```

Şirkette çalışan sayısının departmanlara dağılımını gösteren bir pasta grafiği yapalım:

```python
df_departments['number_of_emp'].plot(
    kind = 'pie',
    figsize = (20,10),
    startangle = 90,
    autopct = '%1.1f%%',
    labels = None,
    shadow = False ,
    pctdistance = 1.1
)

plt.axis('equal')
plt.legend(labels = df_departments.index, loc = 'upper left', prop = {'size' : 8})
plt.show()
```

#### Grafik Yorumu:
 Şirkette en fazla çalışan satış departmanında bulunuyor. Sonra bu sırayı IT departmanı izliyor.

Departmanlara göre ortalama memnuniyet katsayısını çubuk grafiği ile gözlemleyelim:


```python
df_departments['satisfaction_level'].plot(
    kind = 'bar' ,
    figsize = (20,10) #Bu boyutlar, grafiği oluşturmak için kullanılan alanın genişliği ve yüksekliği
)
plt.title('Avarage Satisfaction Level By Departments')
plt.ylabel('Average Satisfaction Level')
plt.xlabel('Department')
plt.show()
```

#### Grafik Yorumu:
Oluşturduğumuz grafikte ortalama memnuniyetin departmanlar arasında çok fazla bir değişkenlik göstermediğini gözlemledik. <br> <br>
Ortalama memnuniyetin maaş seviyesi ile değişip değişmediğini gözlemlemek için çizgi grafiği kullanalım:

```python
df_salary_satisfaction = df.groupby(by ='salary')[['satisfaction_level']].mean()
df_salary_satisfaction.plot(
    kind = 'line',
    figsize = (20,10),
    color = 'blue'
)

plt.title("Average Satisfaction By Salary Level")
plt.ylabel('Satisfaction Level')
plt.xlabel('Salary Level')
plt.show()
```

#### Grafik Yorumu:
Oluşturduğumuz grafiğe göre alınan maaş seviyesi ile memnuniyet seviyesi arasında bir ilişki var. Tahmin edilebileceği gibi yüksek maaş alan kategorisindekilerin memnuniyet seviyeleri daha yüksek. Bununla birlikte arada çok büyük bir fark görünmediğinin de altını çizelim:

Şirket çalışanlarının genel memnuniyetlerini yorumlayabilmek için histogram grafiği yapalım, her şeyin yolunda olduğu bir şirkette bu grafiğin sağa dayalı olmasını bekleriz :)

```python
count, bin_edges = np.histogram(df['satisfaction_level'])
print(count)
print(bin_edges)
```

count: Her bir aralığa düşen değerlerin sayısını içeren bir numpy dizisidir. Bu dizi, histogramdaki çubukların yüksekliklerini temsil eder. <br> <br>
bin_edges: Histogramdaki aralıkların sınırlarını içeren bir numpy dizisidir. Bu dizinin uzunluğu, histogramdaki çubuk sayısından bir fazladır. <br> <br>
Bu çıktıya göre, count dizisi her bir aralıktaki değer sayılarını içerir ve bin_edges dizisi histogramdaki aralık sınırlarını temsil eder. Örneğin, ilk aralık 0.09 ile 0.181 arasındaki değerleri içerir ve bu aralığa düşen değer sayısı 1335'tir.

```python
df['satisfaction_level'].plot(
    kind = 'hist',
    figsize = (20,10),
    xticks = bin_edges,
    color = 'b'
)
plt.title('Satisfaction Level Histogram')
plt.ylabel('Number of Employees')
plt.xlabel('Satisfavtion Level')
plt.grid(True)
plt.show()
```

#### Grafik Yorumu:
Oluşturduğumuz histogramdan hareketle bu şirkette çalışanların genelinin memnun olduğunu bununla birlikte memnuniyet katsayısı düşük olan kişilerin de ihmal edilemeyecek kadar fazla (1250 kişiden fazla) olduğunu ve konunun iyileştirmelere açık olduğunu söyleyebiliriz:
